**Q2.** This question provides some practice cleaning variables which have common problems.
1. Numeric variable: For `./data/airbnb_hw.csv`, clean the `Price` variable as well as you can, and explain the choices you make. How many missing values do you end up with? (Hint: What happens to the formatting when a price goes over 999 dollars, say from 675 to 1,112?)
2. Categorical variable: For the `./data/sharks.csv` data covered in the lecture, clean the "Type" variable as well as you can, and explain the choices you make.
3. Dummy variable: For the pretrial data covered in the lecture, clean the `WhetherDefendantWasReleasedPretrial` variable as well as you can, and, in particular, replace missing values with `np.nan`.
4. Missing values, not at random: For the pretrial data covered in the lecture, clean the `ImposedSentenceAllChargeInContactEvent` variable as well as you can, and explain the choices you make. (Hint: Look at the `SentenceTypeAllChargesAtConvictionInContactEvent` variable.)

```python
import numpy as np
import pandas as pd
import seaborn as sns

#1: Numeric Variable

df = pd.read_csv('airbnb_hw.csv', low_memory=False)
print( df.shape, '\n')
df.head()
```

```python
#1: Numeric Variable
price = df['Price']
price.unique()
```

When prices go over 999, there is a comma. The commas need to be deleted so that the values are not just objects.

```python
#1: Numeric Variable
price = df['Price']
# We need to replace the commas so that the numbers are integers
price = price.str.replace(',','')
print( price.unique() , '\n')
# We are changing the price to numerics rather than a float
price = pd.to_numeric(price,errors='coerce')
print( price.unique() , '\n')
print( 'Total missing values: ', sum( price.isnull() ) )
```

```python
#2 Categorical Variable

df = pd.read_csv('sharks.csv', low_memory=False)
df['Type'].value_counts()
#make temporary vector to clean, and temporary variable
type = df['Type']
type = type.replace(['Sea Disaster', 'Boat','Boating','Boatomg'],'Watercraft')
type.value_counts()
type = type.replace(['Invalid', 'Questionable','Unconfirmed','Unverified','Under investigation'],np.nan) # All unclean values
type.value_counts()
#now make that variable the cleaned one
df['Type'] = type
#now delete the temporary vector
del type
df['Type'].value_counts()
# Now we can see the likelihood of a shark to kill you in a provoked vs unprovoked situation
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace(['UNKNOWN', 'F','M','2017'],np.nan) #unclean values
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace('y','Y') #unclean values
pd.crosstab(df['Type'],df['Fatal (Y/N)'],normalize='index')
```

```python
#3 Dummy Variable

df = pd.read_csv('pretrial_data.csv', low_memory=False)
released = df['released']
print(released.unique(),'\n')
print(released.value_counts(),'\n')
#9 is unclear
released = released.replace(9,np.nan)
print(released.value_counts(),'\n')
sum(released.isnull())
#There are 31 values that are missing
#Replace with the clean values
df['released'] = released
del released
```

```python
#4 Missing values not at random

length = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']

# ' ':missing
# make into numeric
length = pd.to_numeric(length,errors='coerce')
#create dummy
length_NA = length.isnull()
#9k values missing 23k value
print( np.sum(length_NA),'\n')

print( pd.crosstab(length_NA, type), '\n')
#4 is when charges were dismissed
#replace with 4
length = length.mask( type == 4, 0)
#replace with nan
length = length.mask( type == 9, np.nan)
#make new dummy
length_NA = length.isnull()
print( pd.crosstab(length_NA, type), '\n')
print( np.sum(length_NA),'\n')
#274 are missing now
df['ImposedSentenceAllChargeInContactEvent'] = length
#replace with clean, delete temporary variables
del length, type
```
